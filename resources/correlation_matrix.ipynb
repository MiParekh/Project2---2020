```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
csv1 = "characters_stats.csv"
powerstats_df = pd.read_csv(csv1, encoding="ISO-8859-1")
powerstats_df
```

```python
corr_df = powerstats_df[["Intelligence", "Strength", "Speed", "Durability", "Power", "Combat"]]
corr_matrix = corr_df.corr()
corr_matrix
```

```python
plt.figure(figsize=(15, 10))
sn.heatmap(corr_matrix, annot=True)
png_filepath = "../static/assets/images/corr_matrix.png"
plt.savefig(png_filepath)
plt.show()
```
