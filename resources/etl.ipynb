# Extract and Transform

```python
# Store filepath in a variable

#This csv contains all stats score
csv1 = "characters_stats.csv"
#This csv contains character profiles
csv2 = "marvel_characters_info.csv"
#This csv contains the list of select Marvel and DC characters 
# from boolean chart
csv3 = "HeroPowerResolution.csv"
#This csv contains biography of the superheroes 
csv4 = "biography.csv"


# Read our Data file with the pandas library
# Not every CSV requires an encoding, but be aware this can come up
powerstats_df = pd.read_csv(csv1, encoding="ISO-8859-1")
characters_df = pd.read_csv(csv2, encoding="ISO-8859-1")
rwpowers_df = pd.read_csv(csv3, encoding="ISO-8859-1")
bio_df = pd.read_csv(csv4, encoding="ISO-8859-1", index_col=0)
```

```python
# creates a list of character IDs
character_id = list(range(1,732))

query_url = f"https://superheroapi.com/api/{accesstoken}/"

# set up lists to hold reponse info
record = 1
set_number = 1
name = []
full_name = []
alter_egos = []
aliases = []
place_of_birth = []

# Import time so as not to get blocked
import time

# Loop through the list of cities and perform a request for data on each
for c in character_id:
    response = requests.get(query_url + f"{c}/biography").json()
    try:
#         time.sleep(1)
        name.append(response["name"])
        full_name.append(response["full-name"])
        alter_egos.append(response["alter-egos"])
        aliases.append(response["aliases"])
        place_of_birth.append(response["place-of-birth"])
        print(f"Processing Record {record} of Set {set_number} | {c}")
        record += 1
        set_number += 1
    except KeyError:
        print(f"Character not found | {c}")

print("Done processing...")
```

# Load Data to SQL

```python
# Connect to SQLITE
conn = sqlite3.connect(("superheroes.sqlite"))
```

```python
# Use pandas to load csv converted DataFrame into database
characters_df.to_sql('characters', conn, if_exists='replace', index=False)
powerstats_df.to_sql('powerstats', conn, if_exists='replace', index=False)
rwpowers_df.to_sql('rwpowers', conn, if_exists='replace', index=False)
bio_df.to_sql('bio', conn, if_exists='replace', index=False)
```
